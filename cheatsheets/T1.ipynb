# Limpieza datos

dropna: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html

```python
# Borrando datos nulos

# Eliminar filas donde todas las columnas sean nulas
data.dropna(axis=0, how="all")
```

```python
# Eliminar filas donde alguna de las columnas sean nulas
data.dropna(axis=0, how="any")
```

fillna: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html

```python
# rellenando datos nulos pasando un valor
data['body'].fillna('Desconocido')
```

```python
# rellenado datos nulos mediante algun metodo
data["age"].fillna(method="ffill")
```

```python
# calculando variables dummy
pd.get_dummies(data["sex"], prefix="sex")
```

```python
# Agregando variables al dataset
data = pd.concat([data, pd.get_dummies(data["sex"], prefix="sex")], axis = 1)
data.head(5)
```

# Graficos

plot: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html

```python
data = pd.read_csv("../datasets/customer-churn-model/Customer Churn Model.txt")
```

```python
# Graficos in line
%matplotlib inline 
```

## scatter o dispersion

```python
data.plot(kind="scatter", x="Day Mins", y="Day Charge")
```

```python
# multitples graficos
figure, axs = plt.subplots(2,2, sharey=True, sharex=True)
data.plot(kind="scatter", x="Day Mins", y ="Day Charge", ax=axs[0][0])
data.plot(kind="scatter", x="Night Mins", y="Night Charge", ax=axs[0][1])
data.plot(kind="scatter", x="Day Calls", y ="Day Charge", ax=axs[1][0])
data.plot(kind="scatter", x="Night Calls", y="Night Charge", ax=axs[1][1])
```

## Histograma de frecuencias

```python
# Calculo de cantidad de intervalos
k = int(np.ceil(1+np.log2(3333)))
# grafico
plt.hist(data["Day Calls"], bins = k) #bins = [0,30,60,...,200]
plt.xlabel("NÃºmero de llamadas al dÃ­a")
plt.ylabel("Frecuencia")
plt.title("Histograma del nÃºmero de llamadas al dÃ­a")
```

## Boxplot o diagrama de caja y bigotes

```python
plt.boxplot(data["Day Calls"])
plt.ylabel("NÃºmero de llamadas diarias")
plt.title("Boxplot de las llamadas diarias")
```
